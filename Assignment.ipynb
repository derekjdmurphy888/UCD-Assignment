# import numpy as np
## Log Returns
## Standard Deviation
## Volatility

```python
import numpy as np
```

```python
data["Log returns"] = np.log(data["Close"]/data["Close"].shift())
```

```python
data.head()
```

```python
data.tail()
```

```python
data["Log returns"].std()
```

```python
volatility = data["Log returns"].std()*252**.5
```

```python
volatility
```

```python
str_vol = str(round(volatility,3)*100)
```

```python
str_vol
```

```python
fig, ax = plt.subplots()
data["Log returns"].hist(ax=ax, bins=50, alpha=0.6, color="g")
ax.set_xlabel("Log return")
ax.set_ylabel("Freq of log return")
ax.set_title("SONY Volatility:" + str_vol +"%")
```

## CREATING A MOVING AVERAGE

```python
data.head()
```

## ROLLING & ROLLING MEAN

```python
data["Close"].rolling(2).sum()
```

```python
data["Close"].rolling(2).mean()
```

## MOVING & EXPONENTIAL MOVING AVERAGE

```python
data["MA10"] = data ["Close"].rolling (10).mean()
```

```python
data.tail()
```

```python
data["EMA10"] = data ["Close"].ewm(span =10,adjust = False).mean()
```

```python
data.tail()
```

## MOVING & EXPONENTIAL MOVING AVERAGE  GRAPH

```python
fig, ax = plt.subplots()
data[["MA10", "EMA10"]].loc["2020-01-01":"2021-11-16"].plot(ax=ax)
data["Close"].loc ["2020-01-01":"2021-11-16"].plot(ax=ax, alpha=0.25)
```

```python
expl = data["Close"].ewm(span=12, adjust=False).mean()
exp2 = data["Close"].ewm(span=26, adjust=False).mean()
data["MACD"] = expl - exp2
data["Signal line"] =data ["MACD"].ewm(span=9, adjust=False).mean()
```

```python
data.tail()
```

# MOVING AVERAGE CONVERGENCE DIVERGENCE & SIGNAL LINE GRAPH

```python
fig,ax = plt.subplots()
data[["MACD", "Signal line"]].plot (ax=ax)
data["Close"].plot(ax=ax, alpha=0.25, secondary_y = True)
```

## STOCHASTIC OSCILLATOR INDICATOR

```python
high14 = data["High"].rolling(14).max()
low14 = data["Low"].rolling(14).min()
data["%K"] = (data["Close"] - low14)*100/(high14 - low14)
data["%D"] = data ["%K"].rolling(3).mean()
```

```python
data.tail()
```

## STOCHASTIC OSCILLATOR INDICATOR GRAPH

```python
fig, ax =plt.subplots()
data[["%K", "%D"]].loc["2021-09-01":].plot(ax=ax)
ax.axhline(80, c="r", alpha=0.3)
ax.axhline(20, c="r", alpha=0.3)
data["Close"].loc["2021-09-01":].plot(ax=ax, alpha=0.3, secondary_y=True)
```
